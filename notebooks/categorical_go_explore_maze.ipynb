# Categorical Go Explore

In this notebook, we use Go-Explore with cell defined in the categorical latent space.

```python
import gym
import gym_continuous_maze

env = gym.make("ContinuousMaze-v0")
```

```python
env.reset()
for _ in range(100):
    env.step(env.action_space.sample())
```

```python
from IPython.display import display
from PIL import Image

img = env.render("rgb_array")
display(Image.fromarray(img))
```

```python
from go_explore.categorical_go_explore import GoExploreCatVAE
from stable_baselines3 import SAC

go_explore = GoExploreCatVAE(SAC, "ContinuousMaze-v0", verbose=1)
go_explore.explore(20_000)
```

```python
img = go_explore.model.env.render("rgb_array")
display(Image.fromarray(img))
```
